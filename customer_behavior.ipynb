```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df=df.rename(columns={'ï»¿customer_id':'customer_id'})
```

```python
# create a age group column
labels=['young adults', 'adults', 'middle age', 'senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
#craeting a frequncy table becase out frequecy tabe is in the  text we converting that to the  numbers
frequently={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
df['frequently_purchased_date']=df['frequency_of_purchases'].map(frequently)
```
